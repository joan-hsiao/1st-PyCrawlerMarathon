```python
import os
path = r'C:\Users\paran\OneDrive - 大聯大投資控股股份有限公司\Job\Z-Other\PPT_Sample\Icons'

# 獲取該目錄下所有檔案，存入列表中
f = os.listdir(path)
print(len(f))

n = 0
i = 0
for i in f:
    # 設定舊檔名（就是路徑+檔名）
    oldname = f[n]
    oldnamearr = oldname.split('-', 1)      
    
    # 設定新檔名
    newnamearr = oldnamearr[1].split('.', 1)
    newname = newnamearr[0] + '.' + newnamearr[1]
    
    # 用os模組中的rename方法對檔案改名
    if not os.path.isfile(path + '\\' + newname):
        os.rename(path + '\\' + oldname, path + '\\' + newname)
        print(oldname, '======>', newname)
    else:
        newname = newnamearr[0] + "-copy" + '.' + newnamearr[1]
        os.rename(path + '\\' + oldname, path + '\\' + newname)
        print(oldname, '======>', newname)

    n += 1
```
